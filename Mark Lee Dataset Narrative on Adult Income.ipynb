In the below plot, let's see the counts of our data in terms of each individual working class type

As can be easily discerned, the private worktype is the most popular work type in this data. 

```python
df.workclass.value_counts().plot.bar()
plt.ylabel('Counts')
plt.xticks(rotation='45')
plt.title('Count of Workclass Type')
```

However, to see each worktype's relationship with income levels, we need to see the proportion of income type within each work type. As can be seen below

```python
plt.figure(figsize=(10, 5))  # Adjust the figure size as needed

# Calculate the proportions for each category in the 'race' column
proportions = df.groupby(['workclass', 'income']).size() / df.groupby('workclass').size()
proportions = proportions.reset_index(name='proportion')


# Create the bar plot using sns.barplot
sns.barplot(x='workclass', y='proportion', hue='income', data=proportions)

# Optionally, you can rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Labeling the axes and adding a title
plt.xlabel('Workclass')
plt.ylabel('Proportion')
plt.title('Proportion of High vs Low Income Earners by Workclass')
```

The highest proportion of higher income earners (over 50k) resides within the self-employed group followed by federal-gov employment.

Let's continue this type of analysis for every feature in this data.

Let's investigate Race vs Income

- Of most notable, Asian-Pacific Islanders exhibit the highest proportion of high earners
- This is followed closely by White
- Every other race falls behind dramatically in the high income category

These findings actually complicate the current cultural zeitgeist (and conventional wisdom) of Critical Race theory that purports that systematic racism has created these socio-economic realities. 

- On one hand, yes, African-Americans fall far behind Whites in terms of income outcomes. 
- However, Asian-Americans actually top Whites in the same survey towards income, and it can be argued that Asian-Americans since the end of slavery suffered similar levels of systemic oppression. 

Academic literature is replete with different frameworks that attempt to deconstruct these social realities, but it is the author's opinion that both liberal and conservative ideologies that dissect these different success outcomes are reductionistic (perhaps purposely so) to frame reality towards their political agendas. 

There are a lot of sociological evidences to support either side, but nonetheless, this tables insights does provide evidence that the truth seems to be much more complicated vs general public perception.


```python
plt.figure(figsize=(10, 5))  # Adjust the figure size as needed

# Calculate the proportions for each category in the 'race' column
proportions = df.groupby(['race', 'income']).size() / df.groupby('race').size()
proportions = proportions.reset_index(name='proportion')

# Create the bar plot using sns.barplot
sns.barplot(x='race', y='proportion', hue='income', data=proportions)

# Optionally, you can rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Labeling the axes and adding a title
plt.xlabel('Race')
plt.ylabel('Proportion')
plt.title('Proportion of Income by Race')
```
